```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# 2. LOAD DATASET
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Marwadi_uiniversity/SEM-II/ML/Matereal/knn_dataset.csv')
df.head()
```

```python
# Convert target variable (Yes/No â†’ 1/0)
df['Job_offered'] = df['Job_offered'].map({'Yes': 1, 'No': 0})
```

```python
# One-Hot Encoding for categorical features
X = pd.get_dummies(df.drop('Job_offered', axis=1))
y = df['Job_offered']
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)
```

```python
# Apply KNN for different K values
for k in range(1,5):

    classifier = KNeighborsClassifier(n_neighbors=k)

    classifier.fit(X_train, y_train)

    y_pred = classifier.predict(X_test)

    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix for k =", k)
    print(cm)

    accuracy = accuracy_score(y_test, y_pred)*100
    print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
    print("-----------------------------------")
```
